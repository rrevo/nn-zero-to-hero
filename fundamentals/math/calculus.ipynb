## Exponents and Logarithms

$a^x=b$

base - a \
exponent - x \
value - b 

$log_{a}(b)=x$



| Operator | Exponent property | Logarithm property |
| --- | -- | -- |
| Multiplication | $x^m \times x^n = x^{m+n}$  | $log(a \times b) = log(a) + log(b)$ |
| Division | $\frac {x^m} {x^n} = x^{m-n}$ | $ log(\frac {a} {b}) = log(a) - log(b) $ |
| Exponentiation | $(x^m)^n = x^{mn}$ | $ log(a^n) = n \times log(a)$ |
| Zero Exponent | $ x^0 = 1 $ | $ log(1) = 0 $ |
| Inverse | $x^{-1} = \frac 1 x $ | $ log(x^{-1} = log(\frac 1 x) = - log(x)$ |

## Derivatives

Derivative is a slope of a function at a particular point.\
It measures the rate of change at that point in the function. \

Visually it is the slope or tangent at a point in the function. \
Slope is approximately calculated by delta of rise over run for two close points at slope.\

For single variable,
$m=\frac {y_2 - y_1} {x_2 - x_1}$

Single variable

$$
f(x)=x^2 \\
\frac{\partial }{\partial x} f(x) = \frac{\partial }{\partial x} x^2 = 2x
$$

```python
x = symbols('x')
f = x**2
dx_f = diff(f, x)
print(dx_f)
```

Two variables

$$
f(x, y) = 2x^3 + 3y^3 \\
\frac{\partial }{\partial x} 2x^3 + 3y^3 = 6x^2 \\
\frac{\partial }{\partial y} 2x^3 + 3y^3 = 9y^2 \\
$$

```python
x, y = symbols('x y')
f = 2*x**3 + 3*y**3

dx_f = diff(f, x)
dy_f = diff(f, y)

plot3d(f)
```

## Derivative Chain rule

Given
$y=f(x)$, and $z=f(y)$,
derivative of z with respect to x is 
$$
\frac{\partial }{\partial x} z = \frac{\partial }{\partial y} z \times \frac{\partial }{\partial x} y
$$

## Integral

Inverse of a derivative is an integral. It signifies the area under a curve.


```python
x = symbols('x')
f = x**2 + 1
area = integrate(f, (x, 0, 1))
print(area)
```
